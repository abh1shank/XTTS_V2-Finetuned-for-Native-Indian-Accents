```python
import os
import torch
from scipy.io.wavfile import write
from TTS.tts.configs.xtts_config import XttsConfig
from TTS.tts.models.xtts import Xtts
```

```python
CONFIG_PATH = "/content/drive/MyDrive/XTTS_checkpoints/GPT_XTTS_v2.0_LJSpeech_FT-August-07-2024_10+00PM-0000000/config.json"
XTTS_CHECKPOINT = "/content/drive/MyDrive/XTTS_checkpoints/GPT_XTTS_v2.0_LJSpeech_FT-August-07-2024_10+00PM-0000000/model.pth"
TOKENIZER_PATH = "/content/drive/MyDrive/XTTS_checkpoints/XTTS_v2.0_original_model_files/vocab.json"
SPEAKER_REFERENCE="/content/train_assamesemale_01588.wav"
```

```python
OUTPUT_WAV_PATH = "xtts-ft.wav"

print("Loading model...")
config = XttsConfig()
config.load_json(CONFIG_PATH)
model = Xtts.init_from_config(config)
model.load_checkpoint(config, checkpoint_path=XTTS_CHECKPOINT, vocab_path=TOKENIZER_PATH, use_deepspeed=False)
model.cuda()

print("Computing speaker latents...")

print("Inference...")
```

```python
gpt_cond_latent, speaker_embedding = model.get_conditioning_latents(audio_path=[SPEAKER_REFERENCE])
```

```python
import torchaudio
```

```python
out = model.inference(
"Look, I have always believed in the power of our democratic process. However, it's no secret that there were some irregularities in the last election.",
"en",
gpt_cond_latent,
speaker_embedding,
temperature=0.7, # Add custom parameters here
)
torchaudio.save(OUTPUT_WAV_PATH, torch.tensor(out["wav"]).unsqueeze(0), 24000)
```
