```python
import os
import shutil
import random

def is_english_audio(file_path):
    return 'english' in file_path.lower()

def convert_to_ljspeech(transcript_file, output_dir, num_samples=10000):
    os.makedirs(output_dir, exist_ok=True)

    with open(transcript_file, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    english_lines = [line.strip().split('|') for line in lines if len(line.strip().split('|')) == 2 and is_english_audio(line.strip().split('|')[0])]

    selected_lines = random.sample(english_lines, min(num_samples, len(english_lines)))

    metadata = []
    for file_path, transcript in selected_lines:
        audio_filename = os.path.basename(file_path)
        new_audio_path = os.path.join(output_dir, 'wavs', audio_filename)

        os.makedirs(os.path.dirname(new_audio_path), exist_ok=True)
        file_path = file_path.replace('/datadrive/', '/content/')
        shutil.copy2(file_path, new_audio_path)

        meta_name = audio_filename.replace('.wav', '')
        metadata.append(f"{meta_name}|{transcript}|{transcript}")

    with open(os.path.join(output_dir, 'metadata.csv'), 'w', encoding='utf-8') as f:
        f.write('\n'.join(metadata))

    print(f"Converted {len(metadata)} files to LJSpeech format.")

transcript_file = '/content/data/transcript_global.txt'
output_dir = '/content/LJspeech/'
convert_to_ljspeech(transcript_file, output_dir, num_samples=10000)
```

```python
folder_path = '/content/LJspeech'
zip_file_path = '/content/LJspeech.zip'
drive_path = '/content/drive/MyDrive/LJspeech.zip'
shutil.make_archive('/content/LJspeech', 'zip', folder_path)
shutil.move(zip_file_path, drive_path)
print(f"Zip file has been created and moved to Google Drive: {drive_path}")
```
